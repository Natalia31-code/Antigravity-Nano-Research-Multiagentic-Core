## 1. Panorama general

La nanotoxicología busca entender cómo las propiedades fisicoquímicas de una nanopartícula influyen en su efecto biológico o ambiental. En el contexto de Machine Learning, el problema suele formularse como:

- clasificación binaria: tóxica / no tóxica
- clasificación multiclase: niveles de riesgo
- regresión: score continuo de toxicidad, IC50, LC50 o una métrica equivalente

La tendencia actual es usar descriptores fisicoquímicos, modelos supervisados y validación cruzada para predecir toxicidad con menor costo experimental.

## 3. Datasets públicos sobre toxicidad de nanopartículas

### Fuentes principales recomendadas
1. **eNanoMapper (eNM)**
   - Portal abierto con datos sobre propiedades fisicoquímicas y efectos biológicos de nanomateriales.
   - Muy útil para clasificación o regresión de toxicidad.
   - Excelente complemento para el proyecto.

2. **ModNano**
   - Base curada de descriptores y ensayos de toxicidad.
   - Buena opción si se busca una estructura más compacta y limpia.

3. **NanoHUB / curated nanotoxicity resources**
   - Plataforma con herramientas y recursos útiles para simulación, datos y educación.
   - Sirve para contextualizar y enriquecer el reporte.

4. **CEINT / compendios curados de literatura**
   - Datasets derivados de revisión bibliográfica y experimentos curados.
   - Muy útiles cuando el objetivo es justificar toxicidad con endpoints experimentales.

5. **Tox21 como fuente complementaria**
   - No es un dataset puro de nanopartículas, pero puede ayudar en comparación metodológica o transfer learning.

### Recomendación práctica para este proyecto
- Usar eNanoMapper como fuente principal para labels de toxicidad.
- Conservar el dataset del curso como base estructural de descriptores.
- Si es necesario, complementar con un subconjunto más pequeño y curado.

## 4. Variables más utilizadas en nanotoxicología

Las variables que más aparecen en modelos predictivos de toxicidad de nanopartículas son:

- tamaño de partícula
- forma / geometría
- composición química
- carga superficial / potencial zeta
- área superficial específica
- fracción superficial / reactividad
- estabilidad energética
- agregación / aglomeración
- recubrimiento o funcionalización
- dosis
- tiempo de exposición
- medio experimental
- línea celular o sistema biológico
- endpoint medido: viabilidad, citotoxicidad, ROS, genotoxicidad, IC50, LC50

## 6. Métricas para evaluar modelos predictivos

### Para clasificación binaria
- Accuracy
- Precision
- Recall / Sensitivity
- Specificity
- F1-score
- ROC-AUC

### Para regresión
- MAE
- MSE
- RMSE
- R²

### Métrica recomendada para este proyecto
Si el problema se formula como clasificación binaria, la métrica principal debería ser **F1-score** o **Recall** si el costo de no detectar toxicidad es alto.

### Justificación
En nanotoxicidad, un falso negativo puede ser más costoso que un falso positivo, porque dejar pasar una nanopartícula tóxica tiene impacto experimental y de seguridad.
